# 1. Checking Outliers & Frequency Distribution

### Check for outliers

```python
merged_df.dtypes
```

```python
# Generate box plots of each column, as long as it is not an object

plt.figure(figsize=(30, 60))

for i, col in enumerate(merged_df.columns):
    if merged_df[col].dtype != 'object':  
        ax = plt.subplot(9, 2, i + 1)  
        sns.boxplot(data=merged_df, x=col, ax=ax)
        ax.set_xlabel(col, fontsize=24)
        ax.set_title(f"Boxplot of {col}", fontsize=30)
        plt.xlabel(col)  
plt.tight_layout()  
plt.show()
```

Only Year, Net Income Ratio and Gross Profit Ratio, have most of their respective data points between the interquartile range (between 25%-75% of the total value). This observation is expected since both ratios are a proportion of revenue, which naturally restricts the range.

While the remaining metrics (e.g. Revenue, Cost of Revenue, Operating Expenses) are largely made up of outliers, they cluster on the lower end and potentially indicate a skew toward smaller values.

We can use a histogram later to have a clearer view of the distributions.

```python
for i, col in enumerate(merged_df.columns):
    if merged_df[col].dtype != 'object':
        Q1 = merged_df[col].quantile(0.25)
        Q3 = merged_df[col].quantile(0.75)
        IQR = Q3 - Q1
        # Define outlier thresholds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers
        outliers = merged_df[(merged_df[col] < lower_bound) | (merged_df[col] > upper_bound)]

        print("Outliers:")
        print(outliers[[col]])
        
        # Check the percentage of outliers
        outlier_percentage = len(outliers) / len(merged_df) * 100
        print(f"Percentage of Outliers in {col}: {outlier_percentage:.2f}%")

        # Insert a separator after each column
        print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx\n")
```

```python
# Further reading into Revenue Growth outliers
revenue_growth_Q1 = merged_df['Revenue Growth'].quantile(0.25)
revenue_growth_Q3 = merged_df['Revenue Growth'].quantile(0.75)
revenue_growth_IQR = revenue_growth_Q3 - revenue_growth_Q1
revenue_growth_lower_bound = revenue_growth_Q1 - 1.5 * revenue_growth_IQR
revenue_growth_upper_bound = revenue_growth_Q3 + 1.5 * revenue_growth_IQR

# Identify outliers
revenue_growth_outliers = merged_df[
    (merged_df['Revenue Growth'] < revenue_growth_lower_bound) | 
    (merged_df['Revenue Growth'] > revenue_growth_upper_bound)
]['Revenue Growth']

# Calculate statistics for the outliers
if not revenue_growth_outliers.empty:
    min_outlier = revenue_growth_outliers.min()
    mean_outlier = revenue_growth_outliers.mean()
    max_outlier = revenue_growth_outliers.max()

    print(f"Min Outlier: {min_outlier:.2f}")
    print(f"Mean Outlier: {mean_outlier:.2f}")
    print(f"Max Outlier: {max_outlier:.2f}")
else:
    print("No outliers detected in Revenue Growth.")
```

**Outlier Distribution**:

Most columns have no outliers above 1.5 times of the interquartile range.

Operating Income, Net Income, Operating Expenses and EBITDA show small percentages of outliers and are likely natural variations of financial metrics. This also indicates that the operational metrics are stable and reliable.

**Revenue Growth**:

12.19% of the data in Revenue Growth are identified as outliers. Revenue growth rates can have high variability as a result of market conditions and company expansions.

merged_df is also sorted by Stock & Year to show the year-on-year growth in Revenue more clearly. 

Hence, this does not reflect an issue with the data or affect the consistency of the other metrics in the dataset.

# 2. Analyzing Revenue

### Total Revenue and Revenue Growth Over Time

```python
merged_df.columns
```

```python
# Aggregate total revenue by 'Year'
summary = merged_df.groupby('Year').agg({
    'Revenue': 'sum'
}).reset_index()

# Calculate revenue growth as the percentage change from the previous year
summary['Revenue Growth'] = summary['Revenue'].pct_change() * 100

# Create a combo chart: Revenue as bars and Revenue Growth as a line
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Revenue as bars
ax1.bar(summary['Year'], summary['Revenue'], color='blue', label='Total Revenue', alpha=0.6)
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Total Revenue', fontsize=12, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis for Revenue Growth
ax2 = ax1.twinx()
ax2.plot(summary['Year'], summary['Revenue Growth'], marker='o', color='black', label='Revenue Growth (%)')
ax2.set_ylabel('Revenue Growth (%)', fontsize=12, color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Title and grid
plt.title('Total Revenue and Revenue Growth Over Time', fontsize=14)
ax1.grid(True, linestyle='--', alpha=0.7)

# Label all years on the x-axis
ax1.set_xticks(summary['Year'])
ax1.set_xticklabels(summary['Year'], rotation=45, fontsize=10)

# Display the chart
plt.tight_layout()
plt.show()
```

When tracking total revenue over time, total revenue remained almost unchanged from 2000-2018, which indicates minimal improvement in performance. 

Two signficant spikes were observed from 2018-2019 and 2021-2022. This could be attributed to market recovery as businesses have expanded with more revenue. 

However, 2023-2024 saw a sharp decline, bringing total revenue to just slightly above 2019 levels.

### Overall Top 5 vs Current Top 5 Companies by Revenue (2020-2024)

We are also interested to observe how the Top 5 companies with the highest revenue performed from 2020 to 2024 to observe their potential for growth in the near future.
1. Overall Top 5 Companies by Revenue
2. Current Top 5 Companies by Revenue

If any of the Current top 5 companies are part of the Overall Top 5 in recent years, it suggests that they may have greater potential in the comning years and capable of sustaining growth.

We will also assess the revenue share of the Current Top 5 and the Overall Top 5 to understand their contribution to the total revenue generated in 2020-2024.

```python
# Aggregate total revenue by 'Year' and 'Stock'
df_revenue = merged_df.groupby(['Year', 'Stock']).agg({
    'Revenue': 'sum'
}).reset_index()

# Filter data for the years 2020-2024
df_revenue_filtered = df_revenue[df_revenue['Year'].between(2020, 2024)]

# Calculate the current top 5 companies by revenue in the most recent year (e.g., 2024)
current_top_5 = df_revenue[df_revenue['Year'] == 2024].groupby('Stock')['Revenue'].sum().nlargest(5).index

# Aggregate total revenue by company (Stock) for 2020-2024 to find the overall top 5
overall_top_5 = df_revenue_filtered.groupby('Stock')['Revenue'].sum().nlargest(5).index

# Filter data for the current and overall top 5 companies
df_current_top_5 = df_revenue[df_revenue['Stock'].isin(current_top_5)]
df_overall_top_5 = df_revenue[df_revenue['Stock'].isin(overall_top_5)]
```

```python
# Visualize both current and overall top 5 revenue over time (2020-2024)
plt.figure(figsize=(12, 6))

# Plot current top 5 companies
for stock in current_top_5:
    company_data = df_current_top_5[df_current_top_5['Stock'] == stock]
    plt.plot(company_data['Year'], company_data['Revenue'], marker='o', label=f"Current Top 5 - {stock}")

# Plot overall top 5 companies
for stock in overall_top_5:
    company_data = df_overall_top_5[df_overall_top_5['Stock'] == stock]
    plt.plot(company_data['Year'], company_data['Revenue'], marker='x', label=f"Overall Top 5 - {stock}", linestyle='--')

# Title and labels
plt.title('Comparison of Current Top 5 and Overall Top 5 Revenue (2020-2024)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.legend(title='Company (Stock)', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(ticks=range(2020, 2025))

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()

# Print the revenue share of the Overall Top 5 Companies and Current Top 5 Companies
overall_revenue = df_revenue_filtered['Revenue'].sum()
overall_top_5_revenue = df_revenue[df_revenue['Stock'].isin(overall_top_5)]['Revenue'].sum()
current_top_5_revenue = df_revenue[df_revenue['Stock'].isin(current_top_5)]['Revenue'].sum()

revenue_share_overall_top5 = (overall_top_5_revenue / overall_revenue) * 100
revenue_share_current_top5 = (current_top_5_revenue / overall_revenue) * 100

print(f"Revenue share of Overall Top 5 companies: {revenue_share_overall_top5:.2f}%")
print(f"Revenue share of Current Top 5 companies: {revenue_share_current_top5:.2f}%")
```

**MCK, COST & MSFT** are the Current and Overall Top 5 companies that operated with the most revenue. From 2020-2024, each company in the Current Top 5 grew their revenue steadily with each year, resulting in an overall increase by 2024. Despite a much lower total revenue in the industry in 2024, the contrastingly strong performance by the Overall Top 5 companies suggests a resilient and sustainable growth.

**UNH & BRK-B**, which are part of the Overall Top 5, achieved the highest revenue in 2023 but did not remain in the Current Top 5 in 2024. This notable shift suggests a potential decline in their performance or competitive position.

The revenue share discrepancy between Overall Top 5 and Current Top 5 is not significant as MCK, COST & MSFT are common to both groups. BAC-PL, which showed a sudden surge in revenue in 2024, could potentially close the gap further.

# 4. Analyzing Expenses

### Expense Type Breakdown

We are interested in the expense type distribution (Cost of Revenue, Operating Expenses, Depreciation & Amortization) to observe how they compare across the years.

Cost of Revenue = Revenue - Gross Profit

Operating Expenses = Gross Profit - Operating Income

Depreciation & Amortization = EBITDA - Operating Income

Since most companies from the dataset do not suffer from high debt, their net income largely comes from core business operations and not from financing. Hence, Interest Expense is not included.

```python
# Calculate proportions of each category
expense_proportions = merged_df.groupby('Year')[['Cost of Revenue', 'Operating Expenses', 'Depreciation & Amortization']].sum()
expense_proportions['Total'] = expense_proportions.sum(axis=1)
expense_proportions['Cost of Revenue %'] = expense_proportions['Cost of Revenue'] / expense_proportions['Total'] * 100
expense_proportions['Operating Expenses %'] = expense_proportions['Operating Expenses'] / expense_proportions['Total'] * 100
expense_proportions['Depreciation & Amortization %'] = expense_proportions['Depreciation & Amortization'] / expense_proportions['Total'] * 100

expense_proportions
```

```python
# Prepare the data for plotting
years = expense_proportions.index

plt.figure(figsize=(12, 6))

# Plot each category as a line
plt.plot(years, expense_proportions['Cost of Revenue %'], marker='o', label='Cost of Revenue %', color='blue')
plt.plot(years, expense_proportions['Operating Expenses %'], marker='o', label='Operating Expenses %', color='orange')
plt.plot(years, expense_proportions['Depreciation & Amortization %'], marker='o', label='Depreciation & Amortization %', color='green')

# Add title and labels
plt.title('Expense Distribution Over Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.xticks(years, rotation=45)

# Place the legend outside the plot
plt.legend(title='Expense Categories', fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```

Cost of Revenue is consistently higher than Operating Expenses and Depreciation & Amortization. This means that production & procurement costs (Cost of Revenue) are the most significant expense besides admin, marketing and general expenses (Operating Expenses). 

The low D&A also suggests that these companies do not invest in assets that are vulnerable to depreciation.

From 2019, notable shifts in proportions are observed as Cost of Revenue and Operating Expenses began to fluctuate up and down. D&A proportions are also slightly elevated.

### Gross Profit vs Cost of Revenue as a Proportion of Revenue

Since Cost of Revenue + Gross Profit = Revenue, it is useful to see whether the gross profit margin is healthy versus the cost of revenue.

We divide Cost of Revenue & Gross Profit by a single average Revenue rather than per-revenue, which helps to reduce the year-on-year variability and observe the overall trend over time more consistently.

```python
mean_expense_proportions = merged_df.groupby('Year')[['Revenue', 'Gross Profit', 'Cost of Revenue']].mean()
mean_expense_proportions = mean_expense_proportions.rename(columns={
    'Revenue': 'Average Revenue',
    'Gross Profit': 'Average Gross Profit',
    'Cost of Revenue': 'Average Cost of Revenue'
})
```

```python
mean_expense_proportions
```

```python
merged_df
```

```python
mean_expense_proportions['Average Gross Profit (Proportion of Revenue)'] = mean_expense_proportions['Average Gross Profit'] / mean_expense_proportions['Average Revenue']
mean_expense_proportions['Average Cost of Revenue (Proportion of Revenue)'] = mean_expense_proportions['Average Cost of Revenue'] / mean_expense_proportions['Average Revenue']

mean_expense_proportions
```

```python
plt.figure(figsize=(10, 6))

plt.plot(mean_expense_proportions.index, 
         mean_expense_proportions['Average Gross Profit (Proportion of Revenue)'], 
         label='Average Gross Profit (Proportion of Revenue)', 
         marker='o')
plt.plot(mean_expense_proportions.index, 
         mean_expense_proportions['Average Cost of Revenue (Proportion of Revenue)'], 
         label='Average Cost of Revenue (Proportion of Revenue)', 
         marker='o')

plt.title('Average Gross Profit and Cost of Revenue Proportions Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Proportion of Revenue', fontsize=12)

plt.legend(fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

Overall, the Average Gross Profit manages to stay consistently above the Average Cost of Revenue, indicating high Gross Profit margins.

Average Gross Profit and Average Cost of Revenue appear to move together and remain largely stable until 2019, where slight fluctuations are observed and both metrics share an inverse relationship.

When Average Gross Profit increases (51% to 54%), Average Cost of Revenue drops (46% to 44%) and vice versa.

Companies could consider adopting cost management strategies such as renegotiating supplier contracts or diversify revenue streams to control cost of revenues and further stabilize gross profit margins.
